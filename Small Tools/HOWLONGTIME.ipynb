```python
# Information on the 90:10 system

# Im already tried of calculating the time in pico seconds between frames
n_steps = 250_000_000
dt = 0.02 # pico seconds
nst_out = 20_000 # Time steps between output frames

time_conversion_factor = dt * nst_out # Time in pico seconds between output frames


beginning_frame = int(n_steps / nst_out * 0.75)
ending_frame = int(n_steps / nst_out * 1)

print("#### Time Information ####")
Time_total = ending_frame * time_conversion_factor
print(f"Total time: {Time_total} ps --> {Time_total / 1000000} µs")
n_frames_total = int(n_steps / nst_out)
print(f"Total number of frames: {n_frames_total}")


n_frames = ending_frame - beginning_frame
print("\n#### Selection Information ####")
print(f"Beginning frame: {beginning_frame}, Time: {beginning_frame * time_conversion_factor} ps --> {beginning_frame * time_conversion_factor / 1000000} µs")
print(f"Ending frame: {ending_frame}, Time: {ending_frame * time_conversion_factor} ps --> {ending_frame * time_conversion_factor / 1000000} µs")
print(f"Number of frames selected: {n_frames}")



print("\n#### Time between frames ####")
time_per_frame_ps = Time_total / n_frames
print(f"Time between each frame: {time_per_frame_ps} ps")





```
