### The goal of this notebook is to create a calculator that calculates the amount of ions needed for a specific target concentration 

#### *Note*
Net Charges of your lipids or proteins is given from the itp files, which can then be multiplied up by the number of this molecule to give the net current charge

```python
n_water_beads = 56442 ### CHANGE THIS
n_curr_charge = 0     # How many net charges are currently in the system 
                      # And needs to be accounted for when adding ions


target_conc = 0.15 # mol/L (desired concentration of NaCl in the solution)
water_conc = 55.5 # mol/L (approximate concentration of water in pure water)

# Calculate the molar ratio of NaCl to water
molar_ratio = target_conc / water_conc
n_ion_pairs = molar_ratio * n_water_beads * 4 # Each CG beads contains 4 water molecules
print(f"Number of ion pairs to add: {n_ion_pairs:.2f}")

if n_curr_charge < 0:
    n_pos = int(n_ion_pairs) - n_curr_charge
    n_neg = int(n_ion_pairs) 
elif n_curr_charge > 0:
    n_pos = int(n_ion_pairs) 
    n_neg = int(n_ion_pairs) + n_curr_charge
else:
    n_pos = int(n_ion_pairs)
    n_neg = int(n_ion_pairs)
    
print(f"Number of positive ions to add: {n_pos}")
print(f"Number of negative ions to add: {n_neg}")
```
